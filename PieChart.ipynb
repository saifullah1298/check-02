## Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
```

```python
df=pd.read_csv('jsrt_metadata.csv')
df=df.dropna()
df
```

## Malignant

```python
malignant = df.loc[df["state"] == "malignant"]
malignant=malignant.groupby('diagnosis').count()
plt.pyplot.pie(malignant['x'],shadow=True,autopct='%0.1f%%')
```

## Benign

```python
benign = df.loc[df["state"] == "benign"]
benign=benign.groupby('diagnosis').count()
plt.pyplot.pie(benign['y'],shadow=True,autopct='%1.1f%%')
```
